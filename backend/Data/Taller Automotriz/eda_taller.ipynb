**Observaci贸n  -->**  
El conjunto de datos contiene m煤ltiples columnas relacionadas con la fecha. Por ello, se llevar谩 a cabo un proceso de preprocesamiento en el que se consolidar谩 toda la informaci贸n temporal en una 煤nica columna, correctamente formateada. **Observaci贸n  -->**  
Un aspecto crucial identificado en este an谩lisis es la representaci贸n inconsistente de los valores nulos o faltantes en el conjunto de datos. Aunque en la mayor铆a de los casos se utiliza el car谩cter "-" para representarlos, en otros se emplea directamente `np.nan`. Esta falta de uniformidad dificulta el manejo y an谩lisis de los datos. #### **An谩lisis de Datos Faltantes**

```python
# Replace all instances of '-' with "nan"
df.replace('nan', np.nan, inplace=True)
```

```python

# Calculate the percentage of null values in each column
null_percentage = (df.isnull().sum() / len(df)) * 100

# Filter to include only columns with more than 0% null values
null_percentage_filtered = null_percentage[null_percentage > 0]

# Sort the null_percentage Series from highest to lowest
null_percentage_sorted = null_percentage_filtered.sort_values(ascending=False)

# Create a horizontal bar chart
ax = sns.barplot(x=null_percentage_sorted.values, y=null_percentage_sorted.index, hue=null_percentage_sorted.index, dodge=False, palette='viridis')

# Add percentage in each bar
for bar in ax.patches:
    width = bar.get_width()
    ax.text(width,       # X coordinate of text
            bar.get_y() + bar.get_height() / 2,  # Y coordinate of text
            '{:.2f}%'.format(width),  # Text to display (formatted to two decimal places)
            ha='left',  # Horizontal alignment
            va='center',  # Vertical alignment
            color='black',  # Text color
            fontsize=10)  # Font size

plt.xlabel("Porcentaje de Nulos")
plt.ylabel("Columna")
plt.title("Porcentaje de Valores Nulos en Cada Columna")

# Adding grid with custom style
plt.grid(True, linestyle='--', linewidth=0.5)  # Adding grid with dashed lines and custom line width

plt.show()
```

**Observaci贸n  -->**  
Se identific贸 la presencia de m煤ltiples columnas con una proporci贸n significativa de valores faltantes, lo que sugiere que estas variables tienen poca o ninguna utilidad para el an谩lisis. En caso de que el objetivo sea entrenar un modelo, se recomienda eliminarlas, ya que no aportan informaci贸n relevante al conjunto de datos.  

Sin embargo, dado que a煤n no se ha definido el siguiente paso en el procesamiento de este conjunto de datos, se optar谩 por asignar una categor铆a espec铆fica, denominada "Desconocido", a los valores faltantes.
