**Observaci√≥n üí° -->**  
El conjunto de datos contiene m√∫ltiples columnas relacionadas con la fecha. Por ello, se llevar√° a cabo un proceso de preprocesamiento en el que se consolidar√° toda la informaci√≥n temporal en una √∫nica columna, correctamente formateada. Esta estrategia tiene como objetivo reducir la dimensionalidad del conjunto de datos, eliminar redundancias y preservar toda la informaci√≥n relevante para el an√°lisis.

```python
# Convert the FECHA column to a standard format (YYYY-MM-DD)
df['fecha'] = pd.to_datetime(df['fecha'], format='%d/%m/%Y %H:%M').dt.date

# Drop redundant date-related columns
df = df.drop(columns=['mes', 'mes_nombre'], errors='ignore')
```

#### **Frecuencias y Conteo de los Datos**

```python
# See what are the 10 most frequent values for each of the dataframe columns
for column in df.columns:
    frequency_values = df[column].value_counts().head(10)
    print("Top 10 most frequent values for column '{}':".format(column))
    for index, (value, frequency) in enumerate(frequency_values.items(), start=1):
        print("{:<5} {:<30} {:<10}".format(index, value, frequency))
    print("\n===============================================")
```

**Observaci√≥n üí° -->**  
- La columna `cod_epoca` contiene √∫nicamente un valor, lo que la convierte en una variable constante que no aporta informaci√≥n relevante al conjunto de datos ni al modelo a desarrollar. Por lo tanto, se eliminar√°.

- De manera similar, la columna `epoca` tambi√©n posee un √∫nico valor, "ZAFRA", lo que la hace redundante y sin valor informativo para el an√°lisis. En consecuencia, esta columna ser√° eliminada.

- La columna `porc_utilizacion` presenta una gran cantidad de valores nulos, lo que limita su utilidad en el an√°lisis. Debido a esto, se proceder√° a su eliminaci√≥n.

- Las columnas `semana`, `semana_1` y `cod_semana` son redundantes entre s√≠, ya que todas contienen informaci√≥n equivalente. Por lo tanto, se optar√° por consolidarlas en una √∫nica columna para reducir la dimensionalidad y evitar duplicaci√≥n de datos.

```python
# Drop the 'cod_epoca' column as it contains only one unique value
df.drop(columns=['cod_epoca'], inplace=True)

# Drop the 'epoca' column as it contains only one unique value ("ZAFRA")
df.drop(columns=['epoca'], inplace=True)

# Drop the 'porc_utilizacion' column due to the high number of null values
df.drop(columns=['porc_utilizacion'], inplace=True)

# Drop the redundant columns 'semana' and 'semana_1'
df.drop(columns=['semana', 'semana_1'], inplace=True)
```

**Observaci√≥n üí° -->**  
Un aspecto crucial identificado en este an√°lisis es la representaci√≥n inconsistente de los valores nulos o faltantes en el conjunto de datos. Aunque en la mayor√≠a de los casos se utiliza el car√°cter "-" para representarlos, en otros se emplea directamente `np.nan`. Esta falta de uniformidad dificulta el manejo y an√°lisis de los datos. Por lo tanto, se unificar√° la representaci√≥n de los valores faltantes a un formato est√°ndar para garantizar consistencia en el tratamiento de las variables. Posteriormente, se realizar√° un an√°lisis detallado para calcular el porcentaje de valores faltantes en cada columna, lo que permitir√° tomar decisiones informadas sobre su manejo.

```python
# Replace all instances of '-' with "nan"
df.replace('-', 'nan', inplace=True)
# Replace all instances of np.nan with "nan"
df.replace(np.nan, 'nan', inplace=True)
```

#### **An√°lisis de Datos Faltantes**

```python
# Replace all instances of '-' with "nan"
df.replace('nan', np.nan, inplace=True)
```

```python

# Calculate the percentage of null values in each column
null_percentage = (df.isnull().sum() / len(df)) * 100

# Filter to include only columns with more than 0% null values
null_percentage_filtered = null_percentage[null_percentage > 0]

# Sort the null_percentage Series from highest to lowest
null_percentage_sorted = null_percentage_filtered.sort_values(ascending=False)

# Create a horizontal bar chart
ax = sns.barplot(x=null_percentage_sorted.values, y=null_percentage_sorted.index, hue=null_percentage_sorted.index, dodge=False, palette='viridis')

# Add percentage in each bar
for bar in ax.patches:
    width = bar.get_width()
    ax.text(width,       # X coordinate of text
            bar.get_y() + bar.get_height() / 2,  # Y coordinate of text
            '{:.2f}%'.format(width),  # Text to display (formatted to two decimal places)
            ha='left',  # Horizontal alignment
            va='center',  # Vertical alignment
            color='black',  # Text color
            fontsize=10)  # Font size

plt.xlabel("Porcentaje de Nulos")
plt.ylabel("Columna")
plt.title("Porcentaje de Valores Nulos en Cada Columna")

# Adding grid with custom style
plt.grid(True, linestyle='--', linewidth=0.5)  # Adding grid with dashed lines and custom line width

plt.show()
```

**Observaci√≥n üí° -->**  
Se identific√≥ la presencia de m√∫ltiples columnas con una proporci√≥n significativa de valores faltantes, lo que sugiere que estas variables tienen poca o ninguna utilidad para el an√°lisis. En caso de que el objetivo sea entrenar un modelo, se recomienda eliminarlas, ya que no aportan informaci√≥n relevante al conjunto de datos.  

Sin embargo, dado que a√∫n no se ha definido el siguiente paso en el procesamiento de este conjunto de datos, se optar√° por asignar una categor√≠a espec√≠fica, denominada "Desconocido", a los valores faltantes. Este enfoque garantizar√° la consistencia en el tratamiento de los datos y facilitar√° su manejo en etapas posteriores.

```python
# Replace all np.nan values with "Desconocido"
df.fillna("Desconocido", inplace=True)
```
